```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_splitfrom sklearn.metrics import mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python

df = pd.read_csv("/content/sample_data/appliance_energy.csv")
```

```python
df
```

```python
print(df.head())
```

```python
# Features (independent variable) and target (dependent)
# Independent Variable (Temprature)
x = df[['Temperature (°C)']]

# Dependent Variable (Energy Consumption)
y = df['Energy Consumption (kWh)']

print(x, y)
```

```python
# Import required library
from sklearn.model_selection import train_test_split

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

# Create model
model = LinearRegression()

# Train model
model.fit(x_train, y_train)

# Make prediction
y_pred = model.predict(x_test)

print("Prediction Result:")
print(y_pred)
```

```python
#Calculate the  mean square error
mse = mean_squared_error(y_test,y_pred)
print(f"Mean Square Error: {mse}")

#Calculate R-square value
r2 = r2_score(y_test,y_pred)
print(f"R-square Value: {r2}")
```

```python
#plot the test data and regression line
plt.scatter(x_test, y_test, color='blue', label='Test Data')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Energy Consumption prediction using simple Linear Regression')
plt.show()
```

```python
import joblib

#save model to a file
joblib.dump(model,'appliance_energy.model.pk1')
```

```python

# Create Polynomial Features (degree = 2)
poly = PolynomialFeatures(degree=2)

# Transform training data
x_train_poly = poly.fit_transform(x_train)

# Transform test data
x_test_poly = poly.transform(x_test)

# Create model
model = LinearRegression()

# Train model
model.fit(x_train_poly, y_train)

# Make prediction
y_pred = model.predict(x_test_poly)

print("Polynomial Regression Prediction Result:")
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R2 Score:", r2)
```
