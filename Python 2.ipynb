```python
## Create a class
class Dog:
    species = "canine"   # Class attribute

    def __init__(self, name, age):
        self.name = name   # Instance attribute
        self.age = age     # Instance attribute


# Creating objects of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Dane", 12)

print("Species:", dog1.species)
print("Name:", dog1.name)
print("Age:", dog1.age)


print("Species:", dog2.species)
print("Name:", dog2.name)
print("Age:", dog2.age)

# --- Modifying instance attributes ---
dog1.name = "Charlie"   # Change name of dog1
dog1.age = 4            # Change age of dog1

# --- Modifying class attribute ---
Dog.species = "domestic canine"   # Change species for ALL dogs

# --- After modification ---
print("\nAfter modification:")
print("Dog1 -> Species:", dog1.species, "Name:", dog1.name, "Age:", dog1.age)
print("Dog2 -> Species:", dog2.species, "Name:", dog2.name, "Age:", dog2.age)
```

1. Single inheritance

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} makes a sound")

# Child class
class Dog(Animal):
    def sound(self):
        print(f"{self.name} barks!")

# Usage
d = Dog("Buddy")
d.sound()
```

2. Multiple Inheritance

```python
# Parent class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} makes a sound")

# Parent class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

    def show_owner(self):
        print(f"Owner: {self.owner}")

# Child class inherits from both
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def sound(self):
        print(f"{self.name} barks!")

# Usage
d = Dog("Buddy", "Alice")
d.sound()
d.show_owner()
```

3. Multilevel Inheritance

```python
# Grandparent class
class LivingBeing:
    def breathe(self):
        print("Breathing...")

# Parent class
class Animal(LivingBeing):
    def sound(self):
        print("Animals make sounds")

# Child class
class Dog(Animal):
    def sound(self):
        print("Dog barks!")

# Usage
d = Dog()
d.breathe()   # From LivingBeing
d.sound()     # Overridden in Dog
```

```python
# Polymorphism
# Method Overriding

class Animal:
    def speak(self):
        return "I am an animal"

class Dog(Animal):
    def speak(self):
        return "Woof!"

# Creating object of Dog class
d = Dog()

# Calling speak method
print(d.speak())
```

```python
# Polymorphism
# Operator Overloading

class Dog:
    def __init__(self, age):
        self.age = age

    def __add__(self, other):
        return self.age + other.age

d1 = Dog(5)
d2 = Dog(3)

print(d1 + d2)
```

```python
# Polymorhism
# Method Overiding
class Animal:
    def speak(self):
        return "I am an animal"

class Dog(Animal):
    def speak(self):   # Method Overriding
        return "Woof!"

class Cat(Animal):
    def speak(self):   # Method Overriding
        return "Meow!"

# Polymorphism
a = Animal()
d = Dog()
c = Cat()

print(a.speak())
print(d.speak())
print(c.speak())
```

```python
# Operator Overloading
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):   # Operator Overloading
        return self.value + other.value

n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
```

```python
# Encapsulation

class Dog:
    def __init__(self, name, breed, age):
        self.name = name        # Public attribute
        self._breed = breed     # Protected attribute
        self.__age = age        # Private attribute

    # Public method
    def show_details(self):
        print("Name:", self.name)
        print("Breed:", self._breed)
        print("Age:", self.__age)

    # Getter method (to access private variable)
    def get_age(self):
        return self.__age

    # Setter method (to modify private variable)
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age")


# Creating object
d = Dog("Buddy", "Labrador", 3)

# Accessing public variable
print(d.name)

# Accessing protected variable
print(d._breed)

# Accessing private variable using getter
print(d.get_age())

# Modifying private variable using setter
d.set_age(5)

# Display all details
d.show_details()

```

```python
# Data Abstraction
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


# Child class
class Dog(Animal):

    def sound(self):
        return "Woof!"


class Cat(Animal):

    def sound(self):
        return "Meow!"


# Creating objects
d = Dog()
c = Cat()

print(d.sound())
print(c.sound())
```
